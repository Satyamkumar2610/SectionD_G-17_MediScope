### Cleaning `labevents` DataFrame

```python
# Print shape before cleaning
print(f"Shape of labevents before cleaning: {labevents.shape}")

# 1. Convert CHARTTIME to datetime
labevents['charttime'] = pd.to_datetime(labevents['charttime'])

# 2. Drop rows where both VALUE and VALUENUM are null
initial_rows = labevents.shape[0]
labevents.dropna(subset=['value', 'valuenum'], how='all', inplace=True)
dropped_nulls = initial_rows - labevents.shape[0]
print(f"Dropped {dropped_nulls} rows where both 'value' and 'valuenum' were null.")

# 3. Remove duplicate rows
initial_rows = labevents.shape[0]
labevents.drop_duplicates(inplace=True)
dropped_duplicates = initial_rows - labevents.shape[0]
print(f"Dropped {dropped_duplicates} duplicate rows.")

# 4. Add a new column IS_ABNORMAL = True where FLAG column says 'abnormal'
labevents['is_abnormal'] = (labevents['flag'] == 'abnormal')

# Print shape after cleaning
print(f"Shape of labevents after cleaning: {labevents.shape}")

# Display first few rows and info after cleaning to verify
print('\nFirst 3 rows of cleaned labevents:')
display(labevents.head(3))
print('\nColumn info for cleaned labevents:')
labevents.info()
```

### Cleaning `admissions` DataFrame

```python
print(f"Shape of admissions before cleaning: {admissions.shape}")

# 1. Convert ADMITTIME and DISCHTIME to datetime
admissions['admittime'] = pd.to_datetime(admissions['admittime'])
admissions['dischtime'] = pd.to_datetime(admissions['dischtime'])

# 2. Calculate LOS_DAYS
admissions['los_days'] = (admissions['dischtime'] - admissions['admittime']).dt.days

# 3. Standardize text columns to uppercase (assuming all string columns should be uppercased)
for col in ['admission_type', 'admission_location', 'discharge_location', 'insurance', 'language', 'religion', 'marital_status', 'ethnicity', 'edregtime', 'edouttime', 'diagnosis']:
    if col in admissions.columns:
        admissions[col] = admissions[col].astype(str).str.upper()

print(f"Shape of admissions after cleaning: {admissions.shape}")
print('\nFirst 3 rows of cleaned admissions:')
display(admissions.head(3))
print('\nColumn info for cleaned admissions:')
admissions.info()
```

### Cleaning `patients` DataFrame

```python
print(f"Shape of patients before cleaning: {patients.shape}")

# 1. Convert DOB and DOD to datetime
patients['dob'] = pd.to_datetime(patients['dob'])
patients['dod'] = pd.to_datetime(patients['dod'])

# 2. Standardize GENDER to uppercase
patients['gender'] = patients['gender'].astype(str).str.upper()

# 3. Drop duplicates on SUBJECT_ID
initial_rows = patients.shape[0]
patients.drop_duplicates(subset=['subject_id'], inplace=True)
dropped_duplicates = initial_rows - patients.shape[0]
print(f"Dropped {dropped_duplicates} duplicate rows based on 'subject_id'.")

print(f"Shape of patients after cleaning: {patients.shape}")
print('\nFirst 3 rows of cleaned patients:')
display(patients.head(3))
print('\nColumn info for cleaned patients:')
patients.info()
```

### Cleaning `labitems` DataFrame

```python
print(f"Shape of labitems before cleaning: {labitems.shape}")

# 1. Drop duplicates on ITEMID
initial_rows = labitems.shape[0]
labitems.drop_duplicates(subset=['itemid'], inplace=True)
dropped_duplicates = initial_rows - labitems.shape[0]
print(f"Dropped {dropped_duplicates} duplicate rows based on 'itemid'.")

# 2. Strip and title-case LABEL and CATEGORY columns
for col in ['label', 'category']:
    if col in labitems.columns:
        labitems[col] = labitems[col].astype(str).str.strip().str.title()

print(f"Shape of labitems after cleaning: {labitems.shape}")
print('\nFirst 3 rows of cleaned labitems:')
display(labitems.head(3))
print('\nColumn info for cleaned labitems:')
labitems.info()
```

### Merging DataFrames into `lab_master`

```python
# Start with labevents (primary table)
cleaned_database = labevents.copy()

# Merge labitems on ITEMID (left join)
cleaned_database = pd.merge(cleaned_database, labitems[['itemid', 'label', 'category', 'fluid', 'loinc_code']],
                      on='itemid', how='left', suffixes=('_event', '_item'))

# Merge admissions on SUBJECT_ID and HADM_ID (left join)
cleaned_database = pd.merge(cleaned_database, admissions[['subject_id', 'hadm_id', 'los_days', 'admission_type']],
                      on=['subject_id', 'hadm_id'], how='left', suffixes=('_lab', '_adm'))

# Merge patients on SUBJECT_ID (left join)
cleaned_database = pd.merge(cleaned_database, patients[['subject_id', 'gender', 'dob']],
                      on='subject_id', how='left', suffixes=('_master', '_patient'))

print(f"Final shape of cleaned_database: {cleaned_database.shape}")
print("\nColumns in cleaned_database:")
for col in cleaned_database.columns:
    print(f"- {col}")

print('\nFirst 3 rows of cleaned_database:')
display(cleaned_database.head(3))
```

### Final Checks and Export of `cleaned_database`

```python
# 1. Print percentage of missing values per column
print("\nPercentage of missing values per column in cleaned_database (before new imputation):")
print(cleaned_database.isnull().sum() * 100 / len(cleaned_database))

# 2. Print value counts for GENDER, IS_ABNORMAL, and ADMISSION_TYPE
print("\nValue counts for GENDER:")
print(cleaned_database['gender'].value_counts(dropna=False))

print("\nValue counts for IS_ABNORMAL:")
print(cleaned_database['is_abnormal'].value_counts(dropna=False))

print("\nValue counts for ADMISSION_TYPE:")
print(cleaned_database['admission_type'].value_counts(dropna=False))

# Missing value imputation
cleaned_database['flag'].fillna('normal', inplace=True);
cleaned_database['hadm_id'].fillna(0, inplace=True);
cleaned_database['los_days'].fillna(0, inplace=True);
cleaned_database['admission_type'].fillna('UNKNOWN', inplace=True);
cleaned_database['valueuom'].fillna('not recorded', inplace=True);
# valuenum is left as NaN
cleaned_database['loinc_code'].fillna('UNASSIGNED', inplace=True);

# Print missing value counts again to confirm all are fixed
print("\nPercentage of missing values per column in cleaned_database (after imputation):")
print(cleaned_database.isnull().sum() * 100 / len(cleaned_database))

# Print "Cleaning complete " at the end
print("\nCleaning complete ")
```

### Further Cleaning `cleaned_database`: Imputing `valuenum` from `value`

```python
# Define the mapping from 'value' to 'valuenum'
value_to_num_mapping = {
    'NORMAL': 0,
    'TRACE': 0.5,
    'OCCASIONAL': 0.5,
    '1+': 1,
    '2+': 2,
    '3+': 3,
    '4+': 4
}

# Apply the mapping where 'valuenum' is null
for text_val, num_val in value_to_num_mapping.items():
    mask = cleaned_database['valuenum'].isnull() & (cleaned_database['value'].str.upper() == text_val.upper())
    cleaned_database.loc[mask, 'valuenum'] = num_val

# Fill any remaining nulls in 'valuenum' with -1 (unknown sentinel)
cleaned_database['valuenum'].fillna(-1, inplace=True)

# Print how many are still null after mapping
print(f"\nNumber of nulls in 'valuenum' after mapping: {cleaned_database['valuenum'].isnull().sum()}")

# Print value counts of the mapped values
print("\nValue counts for 'valuenum' after mapping:")
print(cleaned_database['valuenum'].value_counts(dropna=False).sort_index())

# Export as clean_lab_master_v3.csv
output_path_v3 = '/content/clean_lab_master_v3.csv'
cleaned_database.to_csv(output_path_v3, index=False)
print(f"\ncleaned_database exported to {output_path_v3}")

print("\nFurther cleaning of valuenum complete ✅")
```
